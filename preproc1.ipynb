```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
df_players = pd.DataFrame(data)
df_players.head(30)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('Â', ''))
df_players.head(45)
```

```python
df_players.to_csv('t20_csv_files/t20_player_info.csv', index = False)
```

```python
df_batting["batsmanName"] = df_batting["batsmanName"].str.strip().str.replace(r"\(.*\)", "", regex=True).str.replace(r"\s+", " ", regex=True).str.title()
df_bowling["bowlerName"]  = df_bowling["bowlerName"].str.strip().str.replace(r"\(.*\)", "", regex=True).str.replace(r"\s+", " ", regex=True).str.title()
```

```python
def overstoballs(overs):
    overs = float(overs)
    whole = int(overs)
    frac=round((overs-whole)*10)
    ballsbowled=(whole*6)+frac
    return ballsbowled

df_bowling["ballsBowled"] = df_bowling["overs"].apply(overstoballs)
numeric_cols = ["overs", "maiden", "runs", "wickets", "economy", "0s", "4s", "6s", "wides", "noBalls", "ballsBowled"]
for col in numeric_cols:
    df_bowling[col] = pd.to_numeric(df_bowling[col], errors="coerce")

# Convert batting columns to numeric
batting_numeric_cols = ["battingPos", "runs", "balls", "4s", "6s", "SR"]
for col in batting_numeric_cols:
    df_batting[col] = pd.to_numeric(df_batting[col], errors="coerce")
df_bowling
```

```python

df_bowling_agg = (df_bowling.groupby("bowlerName").agg(InningsBowled=("match_id", "nunique"),
                                                      OversBowled = ("overs", "sum"),
                                                      ballsBowled = ("ballsBowled", "sum"),
                                                      RunsConceded = ("runs", "sum"),
                                                      Wickets = ("wickets", "sum"),
                                                      DotBalls = ("0s", "sum"))
                                                .reset_index()
                 )

df_bowling_agg
```

```python
df_batting_agg = (df_batting.groupby("batsmanName").agg(InningsBatted=("match_id", "nunique"),
                                                      Runs = ("runs", "sum"),
                                                      BallsFaced = ("balls", "sum"),
                                                      Fours = ("4s", "sum"),
                                                      Sixes = ("6s", "sum"),
                                                      Dismissals = ("out/not_out", lambda x: (x == "out").sum()),
                                                      BattingPosAvg = ("battingPos", "mean"))
                                                    .reset_index()
                 )
df_batting_agg
```

```python
df_batting_agg["batsmanName"] = df_batting_agg["batsmanName"].str.replace('(c)', '', regex = False)
df_bowling_agg["bowlerName"] = df_bowling_agg["bowlerName"].str.replace('(c)', '', regex = False)
```

```python

df_players["name"] = (
    df_players["name"]
    .str.strip()
    .str.replace(r"\(.*\)", "", regex=True)   
    .str.replace(r"\s+", " ", regex=True)     
    .str.title()                              
)
df_players = df_players.drop_duplicates(subset=["name"], keep="first")

player_summary = pd.merge(
    df_players,         # full player list
    df_batting_agg,
    left_on="name",
    right_on="batsmanName",
    how="left"
)


player_summary = pd.merge(
    player_summary,
    df_bowling_agg,
    left_on="name",
    right_on="bowlerName",
    how="left"
)


numeric_cols = player_summary.select_dtypes(include="number").columns
player_summary[numeric_cols] = player_summary[numeric_cols].fillna(0)

player_summary
```

```python

names_original = set(df_players["name"].str.strip())
names_final = set(player_summary["name"].str.strip())
missing_in_final = names_original - names_final

if missing_in_final:
    print(f"{len(missing_in_final)} names from original are missing in final.")
    print("Examples:", list(missing_in_final)[:15])
else:
    print("All names from the original dataset are present in the final dataset.")
```

```python


player_summary.drop(columns=["batsmanName", "bowlerName"], inplace=True)
player_summary.head()
```

```python
player_summary["name"] = player_summary["name"].str.replace('Â', '', regex=False)
player_summary = player_summary.sort_values(by="name").reset_index(drop=True)
```

```python
player_summary["BattingAverage"] = player_summary.apply(
    lambda x: x["Runs"] / x["Dismissals"] if x["Dismissals"] > 0 else 0,
    axis=1
)

player_summary["StrikeRate"] = player_summary.apply(
    lambda x: (x["Runs"] / x["BallsFaced"]) * 100 if x["BallsFaced"] > 0 else 0,
    axis=1
)

player_summary["BoundaryPercent"] = player_summary.apply(
    lambda x: ((x["Fours"]*4 + x["Sixes"]*6) / x["Runs"] * 100) if x["Runs"] > 0 else 0,
    axis=1
)

player_summary["AvgBallsFaced"] = player_summary.apply(
    lambda x: x["BallsFaced"] / x["InningsBatted"] if x["InningsBatted"] > 0 else 0,
    axis=1
)

# Bowling Average
player_summary["BowlingAverage"] = player_summary.apply(
    lambda x: x["RunsConceded"] / x["Wickets"] if x["Wickets"] > 0 else 0,
    axis=1
)

player_summary["Economy"] = player_summary.apply(
    lambda x: (x["RunsConceded"] / x["OversBowled"]) if x["OversBowled"] > 0 else 0,
    axis=1
)

player_summary["BowlingStrikeRate"] = player_summary.apply(
    lambda x: x["ballsBowled"] / x["Wickets"] if x["Wickets"] > 0 else 0,
    axis=1
)

player_summary["DotBallPercent"] = player_summary.apply(
    lambda x: (x["DotBalls"] / x["ballsBowled"] * 100) if x["ballsBowled"] > 0 else 0,
    axis=1
)
player_summary
```

```python
player_summary.to_csv('t20_csv_files/player_summary.csv', index = False)
```

```python
# === OPENERS (Prioritize: Runs > SR > Average) ===
openers = (
    player_summary[
        (player_summary["BattingAverage"] > 30) &
        (player_summary["StrikeRate"] > 140) &
        (player_summary["InningsBatted"] > 3) &
        (player_summary["BoundaryPercent"] > 50) &
        (player_summary["BattingPosAvg"] < 4)
    ]
    .sort_values(by=["BattingAverage", "StrikeRate", "Runs"], ascending=[False, False, False])
    .head(2)
)

# === MIDDLE ORDER (Prioritize: Runs > BattingAverage > SR) ===
middle_order = (
    player_summary[
        (player_summary["BattingAverage"] > 40) &
        (player_summary["StrikeRate"] > 125) &
        (player_summary["InningsBatted"] > 3) &
        (player_summary["AvgBallsFaced"] > 20) &
        (player_summary["BattingPosAvg"] > 2)
    ]
    .sort_values(by=["StrikeRate", "BattingAverage", "Runs"], ascending=[False, False, False])
    .head(3)
)

# === FINISHER (Prioritize: SR > Runs) ===
finisher = (
    player_summary[
        (player_summary["BattingAverage"] > 25) &
        (player_summary["StrikeRate"] > 130) &
        (player_summary["InningsBatted"] > 3) &
        (player_summary["AvgBallsFaced"] > 12) &
        (player_summary["BattingPosAvg"] > 1) &
        (player_summary["InningsBowled"] > 1)
    ]
    .sort_values(by=["StrikeRate", "Runs"], ascending=[False, False])
    .head(1)
)

# === ALL-ROUNDERS (Prioritize: Wickets > BattingAverage > SR) ===
all_rounders = (
    player_summary[
        (player_summary["BattingAverage"] > 15) &
        (player_summary["StrikeRate"] > 140) &
        (player_summary["InningsBatted"] > 2) &
        (player_summary["BattingPosAvg"] > 4) &
        (player_summary["InningsBowled"] > 2) &
        (player_summary["Economy"] < 7) &
        (player_summary["BowlingStrikeRate"] < 20)
    ]
    .sort_values(by=["Wickets", "BattingAverage", "StrikeRate"], ascending=[False, False, False])
    .head(2)
)

# === FAST BOWLERS (Prioritize: Wickets > Economy > BowlingAverage) ===
fast_bowlers = (
    player_summary[
        (player_summary["InningsBowled"] > 4) &
        (player_summary["Economy"] < 7) &
        (player_summary["BowlingStrikeRate"] < 16) &
        (player_summary["BowlingAverage"] < 20) &
        (player_summary["DotBallPercent"] > 40) &
        (player_summary["bowlingStyle"].str.contains("Fast|Medium", case=False, na=False))
    ]
    .sort_values(by=["Wickets", "Economy", "BowlingAverage"], ascending=[False, True, True])
    .head(3)
)

# === COMBINE FINAL XI ===
final_xi = pd.concat([openers, middle_order, finisher, all_rounders, fast_bowlers])
final_xi = final_xi.reset_index(drop = True)

# Show the final XI
final_xi[
    ["name", "playingRole", "team", "Runs", "BattingAverage", "StrikeRate",
     "Wickets", "Economy", "BowlingAverage", "BowlingStrikeRate", "DotBallPercent"]
]
```

```python
final_xi.to_csv('t20_csv_files/final_11.csv', index = False)
```
